```python
#單一折線圖
ts = pd.Series(np.random.randn(200), index=pd.date_range('1/1/2020', periods=200))
ts = ts.cumsum()
print(ts)
ts.plot()
```

```python
#多個折線圖同時畫出
df = pd.DataFrame(np.random.randn(200, 3), index=pd.date_range('1/1/2020', periods=200), columns=["A","B","C"])
df = df.cumsum()
print(df)
df.plot();
```

```python
#圓餅圖
df = pd.Series(np.random.rand(4), index=["A","B","C","D"],name="title")
print(df)
df.plot.pie()
```

```python
#長條圖
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
print(df)
df.plot.bar();
```

```python
df.plot.bar(stacked=True);
```

```python
#箱型圖
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
print(df)
df.boxplot()
```

```python
#散佈圖
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
print(df)
df.plot.scatter(x='A', y='B')
```

```python
boston_data = pd.read_csv ('boston.csv')
print('boston_data: \n', boston_data)
```

```python
#箱型圖
print('BOX: \n', boston_data.boxplot()) #p.s. figsize 可以調整 figure size!!!
```

```python
#中位數
print('median: \n', boston_data.median())
```

```python
print('betwwen 300 to 400', boston_data.median().between(300, 400))
```

```python
#畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係？
boston_data.plot.scatter (x='NOX', y='DIS')
```

```python
age = 'Baby', 'Child', 'Young', 'Teenager', 'Adult'
counts = [21, 62, 10, 53, 77]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lavender']
```

```python
import matplotlib.pyplot as plt
```

```python
patches, texts = plt.pie(counts, colors = colors, shadow = True, startangle = 90)
plt.legend(patches, age, loc = "best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
